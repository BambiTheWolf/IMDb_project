```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.imdb.com/list/ls064287418/")
page
page.status_code
page.content
soup = BeautifulSoup(page.content, "html.parser")
#print(soup.prettify())
Dates = soup.find_all('span', class_ = "lister-item-year text-muted unbold")
Dates = str(Dates)
Dates = Dates.replace('<span class="lister-item-year text-muted unbold">(','').replace(')</span>','').replace('I) (','').replace(' Video','').replace("[",'').replace("]",'')

Dates = Dates.split(', ')
Dates
DateInts = list(map(int, Dates))
DateInts
```
